```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

# 1. กำหนดชื่อคอลัมน์ที่เป็น Feature (0 ถึง 2047)
# จากข้อมูลของคุณมี 2048 features
feat_cols = [str(i) for i in range(2048)]

# 2. แยก X (Features) และ y (ข้อมูลที่เราอยากรู้: Class และ Position)
X_train = FeaturesVector_Traindf[feat_cols].values
X_test = FeaturesVector_Testdf[feat_cols].values

# สร้าง Label สำหรับการทำนาย (ในที่นี้เราเก็บทั้ง Class และ Position)
# เราจะใช้ index เพื่อไปดึงข้อมูล metadata ภายหลัง
train_meta = FeaturesVector_Traindf[['Sub_class_New', 'Sub_Position_Label', 'Path Full']]
test_meta = FeaturesVector_Testdf[['Sub_class_New', 'Sub_Position_Label', 'Path Full']]
```

```python
from sklearn.neighbors import NearestNeighbors

def evaluate_k_nearest(k_value):
    print(f"\n--- Evaluating K = {k_value} ---")
    
    # ใช้ NearestNeighbors เพื่อหา index ของภาพที่ใกล้ที่สุด
    nn = NearestNeighbors(n_neighbors=k_value, metric='euclidean')
    nn.fit(X_train)
    
    # distances = ระยะห่าง, indices = ลำดับแถวใน X_train ที่ใกล้ที่สุด
    distances, indices = nn.kneighbors(X_test)
    
    results = []
    
    for i in range(len(X_test)):
        # ข้อมูลของภาพ Test ปัจจุบัน
        actual_class = test_meta.iloc[i]['Sub_class_New']
        actual_pos = test_meta.iloc[i]['Sub_Position_Label']
        
        # ข้อมูลของเพื่อนบ้านที่ใกล้ที่สุด (Top 1)
        nearest_idx = indices[i][0]
        pred_class = train_meta.iloc[nearest_idx]['Sub_class_New']
        pred_pos = train_meta.iloc[nearest_idx]['Sub_Position_Label']
        
        # เช็คว่า Top 1 ทายถูกไหม
        is_correct = (actual_class == pred_class)
        
        results.append({
            'Test_Class': actual_class,
            'Test_Pos': actual_pos,
            'Matched_Train_Class': pred_class,
            'Matched_Train_Pos': pred_pos,
            'Correct': is_correct
        })
        
    # คำนวณ Accuracy ของ Top-1 
    res_df = pd.DataFrame(results)
    accuracy = res_df['Correct'].mean() * 100
    print(f"Top-1 Accuracy: {accuracy:.2f}%")
    
    return res_df, indices

# รันสำหรับ K=3 และ K=5
res_k3, idx_k3 = evaluate_k_nearest(3)
res_k5, idx_k5 = evaluate_k_nearest(5)
```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors

# 1. เตรียมโมเดล KNN (K=5)
feat_cols = [str(i) for i in range(2048)]
X_train = FeaturesVector_Traindf[feat_cols].values
X_test = FeaturesVector_Testdf[feat_cols].values

# ใช้ metric='euclidean' ตามมาตรฐานการวัดระยะห่างระหว่าง vector
nn = NearestNeighbors(n_neighbors=5, metric='euclidean')
nn.fit(X_train)

# ดึงทั้งระยะห่าง (distances) และลำดับที่ (indices)
distances, indices = nn.kneighbors(X_test)

# 2. สร้าง List สำหรับเก็บข้อมูล
all_results = []

for i in range(len(X_test)):
    # ข้อมูลพื้นฐานจากฝั่ง Test
    test_row = {
        'Path Crop': FeaturesVector_Testdf.iloc[i]['Path Crop'],
        'Sub_class_New': FeaturesVector_Testdf.iloc[i]['Sub_class_New'],
        'Sub_Position_Label': FeaturesVector_Testdf.iloc[i].get('Sub_Position_Label', FeaturesVector_Testdf.iloc[i]['Sub_Position_New'])
    }
    
    # ข้อมูลจากฝั่ง Train (Top 5)
    top5_indices = indices[i]
    top5_distances = distances[i] # ระยะห่างของ 5 อันดับแรก
    
    match_paths = FeaturesVector_Traindf.iloc[top5_indices]['Path Crop'].tolist()
    match_classes = FeaturesVector_Traindf.iloc[top5_indices]['Sub_class_New'].tolist()
    match_positions = FeaturesVector_Traindf.iloc[top5_indices]['Sub_Position_Label'].tolist()
    
    # เก็บข้อมูล list รวม
    test_row['Path Crop_match_train_all'] = match_paths
    test_row['Sub_class_Unlearn_match_train_all'] = match_classes
    test_row['Sub_Position_Unlearn_match_train_all'] = match_positions
    
    # เพิ่มคอลัมน์แยก Top 1 - 5 (Class, Position, และ Distance)
    for rank in range(5):
        test_row[f'Unlearn_pred_15AB_top{rank+1}'] = match_classes[rank]
        test_row[f'Unlearn_pred_Sub_Position_top{rank+1}'] = match_positions[rank]
        test_row[f'Distance_top{rank+1}'] = top5_distances[rank] # <--- เพิ่มคอลัมน์ Distance ตรงนี้
    
    # เก็บข้อมูลสรุปเป็น List
    test_row['Unlearn_prob_15AB_all'] = match_classes
    test_row['Unlearn_prob_Sub_Position_all'] = match_positions
    test_row['Unlearn_Distance_all'] = top5_distances.tolist() # เก็บแบบ list ไว้ด้วยเพื่อความสะดวก
    
    all_results.append(test_row)

# 3. สร้าง DataFrame
df_results = pd.DataFrame(all_results)

print("บันทึกไฟล์สำเร็จ! พบข้อมูลทั้งหมด", len(df_results), "แถว")
df_results.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
from collections import Counter # สำหรับการ Vote เสียงส่วนใหญ่

# 1. เตรียมโมเดล KNN (K=5)
feat_cols = [str(i) for i in range(2048)]
X_train = FeaturesVector_Traindf[feat_cols].values
X_test = FeaturesVector_Testdf[feat_cols].values

nn = NearestNeighbors(n_neighbors=5, metric='euclidean')
nn.fit(X_train)

distances, indices = nn.kneighbors(X_test)

# 2. สร้าง List สำหรับเก็บข้อมูล
all_results = []

for i in range(len(X_test)):
    test_row = {
        'Path Crop': FeaturesVector_Testdf.iloc[i]['Path Crop'],
        'Sub_class_New': FeaturesVector_Testdf.iloc[i]['Sub_class_New'],
        'Sub_Position_Label': FeaturesVector_Testdf.iloc[i]['Sub_Position_Label']
    }
    
    top5_indices = indices[i]
    top5_distances = distances[i]
    
    match_paths = FeaturesVector_Traindf.iloc[top5_indices]['Path Crop'].tolist()
    match_classes = FeaturesVector_Traindf.iloc[top5_indices]['Sub_class_New'].tolist()
    match_positions = FeaturesVector_Traindf.iloc[top5_indices]['Sub_Position_Label'].tolist()
    
    # --- เพิ่มตรรกะการ Vote (KNN Majority Vote) ---
    # Vote สำหรับ Class
    test_row['KNN_Vote_Result_15AB_top3'] = Counter(match_classes[:3]).most_common(1)[0][0]
    test_row['KNN_Vote_Result_15AB_top5'] = Counter(match_classes[:5]).most_common(1)[0][0]
    
    # Vote สำหรับ Position
    test_row['KNN_Vote_Result_Sub_Position_top3'] = Counter(match_positions[:3]).most_common(1)[0][0]
    test_row['KNN_Vote_Result_Sub_Position_top5'] = Counter(match_positions[:5]).most_common(1)[0][0]
    
    # --- เก็บข้อมูลอื่นๆ ตามโครงสร้างเดิม ---
    test_row['Path Crop_match_train_all'] = match_paths
    test_row['Sub_class_Unlearn_match_train_all'] = match_classes
    test_row['Sub_Position_Unlearn_match_train_all'] = match_positions
    
    for rank in range(5):
        test_row[f'Unlearn_pred_15AB_top{rank+1}'] = match_classes[rank]
        test_row[f'Unlearn_pred_Sub_Position_top{rank+1}'] = match_positions[rank]
        test_row[f'Distance_top{rank+1}'] = top5_distances[rank]
    
    test_row['Unlearn_prob_15AB_all'] = match_classes
    test_row['Unlearn_prob_Sub_Position_all'] = match_positions
    test_row['Unlearn_Distance_all'] = top5_distances.tolist()
    
    all_results.append(test_row)

# 3. สร้าง DataFrame
df_results = pd.DataFrame(all_results)

# 4. บันทึกเป็น CSV
# output_path = './Evaluation_Results_KNN_Vote.csv'
# df_results.to_csv(output_path, index=False)

# print(f"บันทึกไฟล์สำเร็จที่: {output_path}")
print(f"พบข้อมูลทั้งหมด {len(df_results)} แถว")
df_results.head()
```

```python
# 1. นิยาม FixedDropout เพื่อให้ Keras รู้จัก
class FixedDropout(tf.keras.layers.Dropout):
    def _get_noise_shape(self, inputs):
        if self.noise_shape is None:
            return self.noise_shape
        return tuple([shape if shape is not None else tf.shape(inputs)[i] 
                      for i, shape in enumerate(self.noise_shape)])

# 2. ใส่ FixedDropout ลงใน custom_objects ตอนโหลด json
custom_objects = {'FixedDropout': FixedDropout}

# 🍉 load json และดึง custom_objects มาใช้
json_path = '/media/HDD/USAI/Model_Unlearn/ML-unlearn/R2/unfreezeB4-B7/models/modelEffNetB5_ML-unlearn_unfreezeB4-B7-R2.json'
with open(json_path, 'r') as json_file:
    loaded_model_json = json_file.read()

# แก้ไขบรรทัดนี้: เพิ่ม custom_objects
model = model_from_json(loaded_model_json, custom_objects=custom_objects)

# 🍉 load weights ตามปกติ
model.load_weights("/media/HDD/USAI/Model_Unlearn/ML-unlearn/R2/unfreezeB4-B7/models/modelEffNetB5_ML-unlearn_unfreezeB4-B7-R2.weights.h5")
print("Loaded model from disk successfully!")
```

```python
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# 1. เลือกรูปแรกมาทดสอบ
sample_row = df_caa_test.iloc[0]
img_path = sample_row['Path Crop']
actual_class = sample_row['Sub_class_New']

# 2. Preprocess รูป
img = load_img(img_path, target_size=(height, width))
img_arr = img_to_array(img) / 255.0
img_arr = np.expand_dims(img_arr, axis=0)

# 3. สั่ง Predict
preds = model.predict(img_arr, verbose=0)[0]
top5_indices = np.argsort(preds)[-5:][::-1]

# --- ตั้งค่าเกณฑ์ขั้นต่ำ (Threshold) ---
# อะไรที่น้อยกว่า 0.00001 ให้ถือว่าเป็น 0
threshold = 1e-5 

# 4. ตรรกะ Strict Override: หา Class สุดท้ายที่มีค่า > Threshold
last_valid_name = labels[top5_indices[0]] 
for idx in top5_indices:
    if preds[idx] > threshold:  # เช็คด้วยเกณฑ์ขั้นต่ำ
        last_valid_name = labels[idx]
    else:
        break 

# 5. แสดงผลตรวจสอบ
print(f"--- Single Image Verification (Threshold Fixed) ---")
print(f"Path: {img_path}")
print(f"Actual Class: {actual_class}\n")

final_list_for_csv = [] 

for i, idx in enumerate(top5_indices):
    confidence = preds[idx]
    
    # ถ้าค่าความมั่นใจสูงกว่าเกณฑ์
    if confidence > threshold:
        name = labels[idx]
        conf_display = f"{confidence*100:>6.2f}%"
    else:
        # ถ้าต่ำกว่าเกณฑ์ ให้ใช้ชื่อสุดท้ายที่ยังมีค่า (AB02)
        name = last_valid_name
        conf_display = f"{confidence*100:>6.2f}% (Below Threshold -> Overridden)"
        
    final_list_for_csv.append(name)
    print(f"อันดับ {i+1}: {name:<10} | Confidence: {conf_display}")

print("-" * 55)
print(f"ค่าที่จะเก็บใน Unlearn_prob_15AB_all:")
print(final_list_for_csv)
```

```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tqdm import tqdm

# 1. ตั้งค่าเกณฑ์ขั้นต่ำ (Threshold) สำหรับ Floating Point
threshold = 1e-5 

# 2. สร้าง List สำหรับเก็บผลลัพธ์
test_results = []

print(f"Starting prediction for {len(df_caa_test)} images with Strict Override Logic...")

# 3. วนลูปทำนายทุกรูป
for i, row in tqdm(df_caa_test.iterrows(), total=len(df_caa_test)):
    try:
        # เตรียมรูปภาพ
        img_path = row['Path Crop']
        img = load_img(img_path, target_size=(height, width))
        img_arr = img_to_array(img) / 255.0
        img_arr = np.expand_dims(img_arr, axis=0)
        
        # Predict
        preds = model.predict(img_arr, verbose=0)[0]
        
        # ดึง Index ของ Top 5 (เรียงจากมากไปน้อย)
        top5_indices = np.argsort(preds)[-5:][::-1]
        
        # --- ขั้นตอนหา Last Valid Name (ตัวสุดท้ายที่มั่นใจเกิน Threshold) ---
        last_valid_name = labels[top5_indices[0]] 
        for idx in top5_indices:
            if preds[idx] > threshold:
                last_valid_name = labels[idx]
            else:
                break 

        # --- สร้างรายชื่อ Top 5 ตามตรรกะ Override ---
        final_top5_names = []
        for idx in top5_indices:
            if preds[idx] > threshold:
                final_top5_names.append(labels[idx])
            else:
                # ถ้าต่ำกว่าเกณฑ์ ให้ใช้ชื่อสุดท้ายที่ยังมีค่า
                final_top5_names.append(last_valid_name)

        # 4. เก็บข้อมูลลงใน Dictionary
        result_entry = {
            'Path Crop': img_path,
            'Sub_class_New': row['Sub_class_New'],
            'Sub_Position_Label': row.get('Sub_Position_Label', row.get('Sub_Position_New', 'N/A')),
            'Unlearn_pred_15AB_top1': final_top5_names[0],
            'Unlearn_pred_15AB_top2': final_top5_names[1],
            'Unlearn_pred_15AB_top3': final_top5_names[2],
            'Unlearn_pred_15AB_top4': final_top5_names[3],
            'Unlearn_pred_15AB_top5': final_top5_names[4],
            'Unlearn_prob_15AB_all': final_top5_names 
        }
        
        test_results.append(result_entry)
        
    except Exception as e:
        print(f"Error at index {i}: {e}")
        continue

# 5. แปลงเป็น DataFrame และบันทึกไฟล์
df_unlearn_results = pd.DataFrame(test_results)
# output_path = './Model_Unlearn_Prediction_Results_Fixed.csv' # ตั้งชื่อไฟล์ใหม่เพื่อกันความสับสน
# df_unlearn_results.to_csv(output_path, index=False)

print(f"\nเสร็จสมบูรณ์! บันทึกไฟล์ที่: {output_path}")
print(df_unlearn_results.head())
```
