```python
import numpy as np
from tqdm.auto import tqdm # ‡πÅ‡∏™‡∏î‡∏á‡πÅ‡∏ñ‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏Ñ‡∏∑‡∏ö‡∏´‡∏ô‡πâ‡∏≤

# 1. ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô Feature (0-2047)
feature_cols = [str(i) for i in range(2048)]

# 2. ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏≠‡∏±‡∏õ‡πÇ‡∏´‡∏•‡∏î
def upload_to_pinecone(df, batch_size=50):
    print(f"‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏≠‡∏±‡∏õ‡πÇ‡∏´‡∏•‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î {len(df)} ‡πÅ‡∏ñ‡∏ß...")
    
    for i in tqdm(range(0, len(df), batch_size)):
        # ‡∏ï‡∏±‡∏î‡πÅ‡∏ö‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏õ‡πá‡∏ô‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏¢‡πà‡∏≠‡∏¢
        batch_df = df.iloc[i : i + batch_size]
        
        vectors_to_upsert = []
        for idx, row in batch_df.iterrows():
            # ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏ß‡∏Å‡πÄ‡∏ï‡∏≠‡∏£‡πå‡πÄ‡∏õ‡πá‡∏ô List ‡∏Ç‡∏≠‡∏á float
            v_values = row[feature_cols].values.astype(float).tolist()
            
            # ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏° Metadata ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡πÅ‡∏•‡∏∞‡∏Ñ‡∏±‡∏î‡∏Å‡∏£‡∏≠‡∏á
            metadata = {
                "sub_class": str(row["Sub_class_New"]),
                "position": str(row["Sub_Position_Label"])
            }
            
            # ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö Pinecone
            vectors_to_upsert.append({
                "id": str(idx), 
                "values": v_values, 
                "metadata": metadata
            })
            
        # ‡∏™‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ç‡∏∂‡πâ‡∏ô Cloud
        index.upsert(vectors=vectors_to_upsert)

# 3. ‡∏£‡∏±‡∏ô‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏á‡∏≤‡∏ô
upload_to_pinecone(FeaturesVector_Traindf)

print("\n‚ú® ‡∏≠‡∏±‡∏õ‡πÇ‡∏´‡∏•‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢!")
```

```python
from collections import Counter
import pandas as pd
from tqdm.auto import tqdm

# 1. ‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤‡∏û‡∏≤‡∏£‡∏≤‡∏°‡∏¥‡πÄ‡∏ï‡∏≠‡∏£‡πå
feature_cols = [str(i) for i in range(2048)]
top_k = 5  # ‡∏Ñ‡πà‡∏≤ K ‡∏Ç‡∏≠‡∏á KNN
results = []

print(f"‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏Å‡∏≤‡∏£‡∏ó‡∏≥ KNN (K={top_k}) ‡∏ö‡∏ô Vector Database...")

# 2. Loop ‡∏ú‡πà‡∏≤‡∏ô Test Set (‡∏™‡∏°‡∏°‡∏ï‡∏¥‡∏•‡∏≠‡∏á‡∏ó‡∏≥ 100 ‡∏£‡∏π‡∏õ‡πÅ‡∏£‡∏Å‡∏Å‡πà‡∏≠‡∏ô‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏ä‡πá‡∏Ñ‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡πá‡∏ß ‡∏´‡∏£‡∏∑‡∏≠‡∏ó‡∏≥‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡∏Å‡πá‡πÑ‡∏î‡πâ‡∏Ñ‡∏£‡∏±‡∏ö)
for idx, row in tqdm(FeaturesVector_Testdf.iterrows(), total=len(FeaturesVector_Testdf)):
    
    # ‡∏î‡∏∂‡∏á Vector ‡∏Ç‡∏≠‡∏á‡∏£‡∏π‡∏õ Test
    test_vector = row[feature_cols].values.astype(float).tolist()
    
    # 3. Query ‡πÑ‡∏õ‡∏ó‡∏µ‡πà Pinecone
    query_response = index.query(
        vector=test_vector,
        top_k=top_k,
        include_metadata=True
    )
    
    # 4. ‡∏î‡∏∂‡∏á Class ‡∏Ç‡∏≠‡∏á‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ô‡∏ö‡πâ‡∏≤‡∏ô 5 ‡∏Ñ‡∏ô‡∏ó‡∏µ‡πà‡πÉ‡∏Å‡∏•‡πâ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
    neighbor_classes = [match['metadata']['sub_class'] for match in query_response['matches']]
    
    # 5. ‡∏ó‡∏≥ Majority Vote (‡∏´‡∏≤‡πÄ‡∏™‡∏µ‡∏¢‡∏á‡∏™‡πà‡∏ß‡∏ô‡πÉ‡∏´‡∏ç‡πà)
    if neighbor_classes:
        vote_result = Counter(neighbor_classes).most_common(1)[0][0]
    else:
        vote_result = "Unknown"
        
    # ‡πÄ‡∏Å‡πá‡∏ö‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡πÑ‡∏ß‡πâ‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö
    results.append({
        "Actual": row["Sub_class_New"],
        "Predicted": vote_result,
        "Neighbors": neighbor_classes
    })

# 6. ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô DataFrame ‡πÅ‡∏•‡∏∞‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì Accuracy
results_df = pd.DataFrame(results)
accuracy = (results_df['Actual'] == results_df['Predicted']).mean()

print(f"\n‚úÖ ‡∏Å‡∏≤‡∏£‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô!")
print(f"üéØ Accuracy ‡∏Ç‡∏≠‡∏á‡∏£‡∏∞‡∏ö‡∏ö KNN ‡∏ö‡∏ô Pinecone: {accuracy * 100:.2f}%")
```

```python
from collections import Counter
import pandas as pd
from tqdm.auto import tqdm

# 1. ‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤‡∏û‡∏≤‡∏£‡∏≤‡∏°‡∏¥‡πÄ‡∏ï‡∏≠‡∏£‡πå
feature_cols = [str(i) for i in range(2048)]
top_k = 5
results = []

print(f"‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏Å‡∏≤‡∏£‡∏ó‡∏≥ KNN (K={top_k}) ‡∏ö‡∏ô Vector Database...")

# 2. Loop ‡∏ú‡πà‡∏≤‡∏ô Test Set ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
for idx, row in tqdm(FeaturesVector_Testdf.iterrows(), total=len(FeaturesVector_Testdf)):
    
    # ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏° Vector
    test_vector = row[feature_cols].values.astype(float).tolist()
    
    # 3. Query ‡πÑ‡∏õ‡∏ó‡∏µ‡πà Pinecone
    query_response = index.query(
        vector=test_vector,
        top_k=top_k,
        include_metadata=True
    )
    
    # 4. ‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≤‡∏Å‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ô‡∏ö‡πâ‡∏≤‡∏ô (Matches)
    # ‡∏î‡∏∂‡∏á‡∏ó‡∏±‡πâ‡∏á Class ‡πÅ‡∏•‡∏∞ Position ‡∏Ç‡∏≠‡∏á‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ô‡∏ö‡πâ‡∏≤‡∏ô‡∏°‡∏≤‡∏î‡∏π
    neighbor_classes = [match['metadata']['sub_class'] for match in query_response['matches']]
    neighbor_positions = [match['metadata']['position'] for match in query_response['matches']]
    
    # 5. ‡∏ó‡∏≥ Majority Vote ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏´‡∏≤ Class ‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö
    if neighbor_classes:
        vote_result = Counter(neighbor_classes).most_common(1)[0][0]
        # ‡πÅ‡∏ñ‡∏°: ‡∏ó‡∏≥ Majority Vote ‡∏Ç‡∏≠‡∏á Position ‡∏î‡πâ‡∏ß‡∏¢ (‡∏ñ‡πâ‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÄ‡∏ä‡πá‡∏Ñ‡∏Ñ‡∏ß‡∏≤‡∏°‡πÅ‡∏°‡πà‡∏ô‡∏¢‡∏≥‡∏Ç‡∏≠‡∏á‡∏°‡∏∏‡∏°)
        vote_pos_result = Counter(neighbor_positions).most_common(1)[0][0]
    else:
        vote_result = "Unknown"
        vote_pos_result = "Unknown"
        
    # 6. ‡πÄ‡∏Å‡πá‡∏ö‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå (‡πÄ‡∏û‡∏¥‡πà‡∏° Path ‡πÅ‡∏•‡∏∞ Position ‡∏Ç‡∏≠‡∏á‡∏£‡∏π‡∏õ Test)
    results.append({
        "Test_Path": row["Path Full"],         # Path ‡∏£‡∏π‡∏õ Test
        "Actual_Class": row["Sub_class_New"], # ‡∏Ñ‡∏•‡∏≤‡∏™‡∏à‡∏£‡∏¥‡∏á
        "Actual_Pos": row["Sub_Position_Label"], # ‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏à‡∏£‡∏¥‡∏á
        "Predicted_Class": vote_result,        # ‡∏Ñ‡∏•‡∏≤‡∏™‡∏ó‡∏µ‡πà‡∏ó‡∏≤‡∏¢ (‡∏à‡∏≤‡∏Å KNN)
        "Predicted_Pos": vote_pos_result,     # ‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏ó‡∏µ‡πà‡∏ó‡∏≤‡∏¢ (‡∏à‡∏≤‡∏Å KNN)
        "Neighbor_Classes": neighbor_classes,
        "Neighbor_Positions": neighbor_positions
    })

# 7. ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô DataFrame
results_df = pd.DataFrame(results)

# 8. ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì Accuracy
acc_class = (results_df['Actual_Class'] == results_df['Predicted_Class']).mean()
acc_pos = (results_df['Actual_Pos'] == results_df['Predicted_Pos']).mean()

print(f"\n‚úÖ ‡∏Å‡∏≤‡∏£‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô!")
print(f"üéØ Accuracy (Class): {acc_class * 100:.2f}%")
print(f"üéØ Accuracy (Position): {acc_pos * 100:.2f}%")

# ‡πÅ‡∏™‡∏î‡∏á‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ï‡∏≤‡∏£‡∏≤‡∏á
results_df.head()
```
